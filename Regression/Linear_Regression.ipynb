PART 3: PURE PYTHON LOGIC (NO AI / ML)
---------------------------------------

Ab wahi cheez normal Python logic se karte hain
(No AI, no machine learning)

Simple idea:

‚ÄúAverage badlav nikal kar future guess karna‚Äù

```python
# Past data
time = [1, 2, 3, 4]
money = [50, 100, 150, 200]

# Har ghante kitna paisa badh raha hai
increase_per_hour = []

for i in range(1, len(time)):
    change = money[i] - money[i-1]
    increase_per_hour.append(change)

average_increase = sum(increase_per_hour) / len(increase_per_hour)

# Future prediction
future_time = 5
predicted_money = average_increase * future_time

print(predicted_money)
```

Line-by-line bachchon jaise samjho

time ‚Üí input data

money ‚Üí output data

Loop ‚Üí har step ka difference

Average ‚Üí ek common rule

Multiply ‚Üí future ka guess

Yani:
Input ‚Üí Process ‚Üí Output

Bilkul maths + graph jaisa hi
Bas ab computer se karwaya

Simple ML code (sirf concept ke liye)

```python
from sklearn.linear_model import LinearRegression

# Training data
X = [[1], [2], [3], [4]]
y = [50, 100, 150, 200]

model = LinearRegression()
model.fit(X, y)

# Prediction
print(model.predict([[5]]))
```
